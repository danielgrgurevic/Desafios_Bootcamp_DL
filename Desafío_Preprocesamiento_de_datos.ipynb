```python
# Convertir columnas numéricas de object a numérico
df['User Rating'] = pd.to_numeric(df['User Rating'], errors='coerce')
df['Number of Votes'] = pd.to_numeric(df['Number of Votes'], errors='coerce')
df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')
```

```python
import re

# Función para extraer años de un rango
def extract_years(year):
    if isinstance(year, str):
        # Buscar todos los años en el formato (XXXX–XXXX) o XXXX
        matches = re.findall(r'(\d{4})', year)
        if matches:
            start_year = int(matches[0])  # Primer año encontrado
            end_year = int(matches[1]) if len(matches) > 1 else None  # Segundo año si existe
            return start_year, end_year
    return None, None  # Retornar None si no hay años válidos

# Aplicar la función para crear dos nuevas columnas
df[['Start Year', 'End Year']] = df['Year'].apply(lambda x: pd.Series(extract_years(x)))
```

```python
# Convertir las columnas a numéricas
df['Start Year'] = pd.to_numeric(df['Start Year'], errors='coerce')
df['End Year'] = pd.to_numeric(df['End Year'], errors='coerce')
```

```python
# Asegurarte de que las columnas sean de tipo float
df['Start Year'] = df['Start Year'].astype(float)
df['End Year'] = df['End Year'].astype(float)
```

```python
df.info()
```

```python
# Rellenar nulos con la media
df['Start Year'].fillna(df['Start Year'].mean(), inplace=True)
df['End Year'].fillna(df['End Year'].mean(), inplace=True)
```

```python
df.head()
```

```python
# Calcular la diferencia entre años
df['Year Difference'] = df['End Year'] - df['Start Year']
df['Year Difference'] = df['Year Difference'].astype(float)
df['Year Difference'].fillna(df['Year Difference'].mean(), inplace=True)
```

```python
from scipy.stats import shapiro

# Verificar la distribución normal usando la prueba de Shapiro-Wilk
normal_columns = []
non_normal_columns = []

for column in all_columns:
    if df[column].dtype in ['float64', 'int64']:
        stat, p_value = shapiro(df[column])
        if p_value > 0.05:  # p > 0.05 significa que no podemos rechazar la hipótesis nula (distribución normal)
            normal_columns.append(column)
        else:
            non_normal_columns.append(column)

print("Columnas con distribución normal:", normal_columns)
print("Columnas con distribución no normal:", non_normal_columns)
```

```python
import numpy as np
import pandas as pd
from scipy import stats
from sklearn.preprocessing import PowerTransformer

# Función para probar las diferentes transformaciones y calcular el p-valor de Shapiro-Wilk
def evaluate_transformations(df, columns):
    best_transformations = {}

    # Recorrer cada columna
    for column in columns:
        data = df[column].dropna()  # Eliminar valores nulos
        transformations = {
            'original': data,
            'logarithmic': np.log1p(data),  # log(1 + x) para evitar log(0)
            'sqrt': np.sqrt(data),
            'cubic_root': np.cbrt(data),
            'inverse': 1 / (data + 1e-6),  # Evitar divisiones por cero
            'boxcox': stats.boxcox(data + 1e-6)[0] if (data > 0).all() else None,  # Solo para datos positivos
            'yeo-johnson': PowerTransformer(method='yeo-johnson').fit_transform(data.values.reshape(-1, 1)).flatten()
        }

        best_pvalue = -1
        best_transformation = None

        # Evaluar cada transformación
        for name, transformed_data in transformations.items():
            if transformed_data is not None:  # Verificar si la transformación es aplicable
                stat, pvalue = stats.shapiro(transformed_data)
                if pvalue > best_pvalue:
                    best_pvalue = pvalue
                    best_transformation = name

        # Guardar el mejor resultado para la columna
        best_transformations[column] = best_transformation
        print(f"Best transformation for {column}: {best_transformation} (p-value: {best_pvalue})")

    return best_transformations

# Lista de columnas de tu DataFrame df que deseas transformar
columns_to_transform = ['Title', 'Genre', 'User Rating', 'Number of Votes', 'Runtime',
               'Summary', 'Stars', 'Certificate', 'Gross', 'Episode',
               'Episode Title', 'Start Year', 'End Year', 'Year Difference']

# Evaluar las transformaciones en las columnas seleccionadas
best_transformations = evaluate_transformations(df, columns_to_transform)
```

```python
# Aplicar la mejor transformación a cada columna
for column in columns_to_transform:
    if best_transformations[column] == 'logarithmic':
        df[column] = np.log1p(df[column])
    elif best_transformations[column] == 'sqrt':
        df[column] = np.sqrt(df[column])
    elif best_transformations[column] == 'cubic_root':
        df[column] = np.cbrt(df[column])
    elif best_transformations[column] == 'inverse':
        df[column] = 1 / (df[column] + 1e-6)
    elif best_transformations[column] == 'boxcox':
        df[column] = stats.boxcox(df[column] + 1e-6)[0]
    elif best_transformations[column] == 'yeo-johnson':
        df[column] = PowerTransformer(method='yeo-johnson').fit_transform(df[column].values.reshape(-1, 1)).flatten()
```

```python
from scipy import stats

# Función para eliminar outliers usando Z-score para datos normales
def remove_outliers_zscore(df, columns, threshold=3):
    for column in columns:
        z_scores = np.abs(stats.zscore(df[column]))
        df = df[z_scores < threshold]  # Mantener solo los valores dentro del umbral de Z-score
    return df

# Función para eliminar outliers usando IQR para datos no normales
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]  # Mantener solo valores dentro de los límites
    return df
```

```python
# Determinar qué columnas tienen distribución normal y cuáles no aproximadamente de manera visual, no se considera episode al ser solo dos valores

normal_columns = ['User Rating', 'Number of Votes', 'Runtime',
               'Summary', 'Stars', 'Gross', 'Start Year', 'End Year', 'Year Difference']

non_normal_columns = ['Title', 'Genre', 'Certificate', 'Episode',
               'Episode Title']


# Eliminar los outliers
df_normal_cleaned = remove_outliers_zscore(df, normal_columns)  # Usar Z-score para columnas normales
df_final_cleaned = remove_outliers_iqr(df_normal_cleaned, non_normal_columns)  # Usar IQR para columnas no normales

# Mostrar el DataFrame después de eliminar los atípicos
print(df_final_cleaned)
```
