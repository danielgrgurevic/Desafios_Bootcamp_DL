**Trabajando con earnings**

Teniendo en cuenta la media poblacional de **21147.29**

1. **Prueba para \( \mu = 20,000 \) (bilateral):**
   - Se rechaza la hipótesis nula para \(\alpha\) de 0.09 a 0.05, lo que indica que la media de la muestra es significativamente diferente de 20,000 con niveles de confianza del 91%, 93%, y 95%.
   - Se acepta la hipótesis nula para \(\alpha\) de 0.03 y 0.01, indicando que con niveles de confianza del 97% y 99%, no hay suficiente evidencia para concluir que la media de la muestra es diferente de 20,000.
   - Dado que la media poblacional es 21,147.29, que está más cerca de 20,000 que de 35,000 o 25,000, es razonable que en niveles de confianza más altos (97% y 99%), la prueba no tenga suficiente poder para rechazar \(H_0\).

2. **Prueba para \( \mu = 35,000 \) (bilateral):**
   - En todos los niveles de significancia, se rechaza la hipótesis nula. Esto sugiere una fuerte evidencia de que la media de la muestra no es 35,000. Esto es consistente con la media poblacional de 21,147.29, que está muy por debajo de 35,000.

3. **Prueba para \( \mu > 25,000 \) (unilateral):**
   - En todos los niveles de significancia, se acepta la hipótesis nula. Esto indica que no hay suficiente evidencia para concluir que la media de la muestra es mayor a 25,000, lo cual tiene sentido ya que la media poblacional es 21,147.29, que es menor a 25,000.

4. **Prueba para \( \mu < 32,000 \) (unilateral):**
   - En todos los niveles de significancia, se rechaza la hipótesis nula. Esto sugiere que hay suficiente evidencia para concluir que la media de la muestra es menor a 32,000, lo cual es coherente con la media poblacional de 21,147.29.


```python
def hypothesis_test(sample, null_value, alpha, alternative):
    t, p = stats.ttest_1samp(sample, null_value, alternative=alternative)

    if p < alpha:
        return f"Con nivel de confianza del {100 * (1 - alpha)}%, se rechaza la hipótesis nula. t es: {t}, p es: {p}"
    else:
        return f"Con nivel de confianza del {100 * (1 - alpha)}%, se acepta la hipótesis nula. t es: {t}, p es: {p}"

sample_earn = df['earn'].sample(n=25, random_state=1)
print(hypothesis_test(sample_earn, 20000, 0.05, 'two-sided'))
print(hypothesis_test(sample_earn, 35000, 0.05, 'two-sided'))
print(hypothesis_test(sample_earn, 25000, 0.05, 'greater'))
print(hypothesis_test(sample_earn, 32000, 0.05, 'less'))
```

```python
# variando los alphas para tener 5 pruebas por hipótesis

def hypothesis_test(sample, null_value, alpha, alternative):
    t, p = stats.ttest_1samp(sample, null_value, alternative=alternative)

    if p < alpha:
        return f"Con nivel de confianza del {100 * (1 - alpha)}%, se rechaza la hipótesis nula. t es: {t}, p es: {p}"
    else:
        return f"Con nivel de confianza del {100 * (1 - alpha)}%, se acepta la hipótesis nula. t es: {t}, p es: {p}"

alphas = [0.09, 0.07, 0.05, 0.03, 0.01]   # menos a 0.1

for alpha in alphas:
    sample_earn = df['earn'].sample(n=25, random_state=1)
    print(hypothesis_test(sample_earn, 20000, alpha, 'two-sided'))
    print(hypothesis_test(sample_earn, 35000, alpha, 'two-sided'))
    print(hypothesis_test(sample_earn, 25000, alpha, 'greater'))
    print(hypothesis_test(sample_earn, 32000, alpha, 'less'))
```

```python
import pandas as pd
import statsmodels.api as sm

def calcular_proporcion_hombres(df, niveles_significancia):
    n = len(df)
    num_hombres = df[df["male"] == 1]["male"].count()
    proporcion_muestral = num_hombres / n

    for alpha in niveles_significancia:
        intervalo_confianza = sm.stats.proportion_confint(num_hombres, n, alpha=alpha)
        print(f"Nivel de significancia: {alpha}")
        print(f"Proporción muestral de hombres: {proporcion_muestral:.2f}")
        print(f"Intervalo de confianza: ({intervalo_confianza[0]:.2f}, {intervalo_confianza[1]:.2f})")
        print()

niveles_significancia = [0.01, 0.05]
calcular_proporcion_hombres(df, niveles_significancia)
```

```python
# Filtrar los datos por género
earn_male = df[df['male'] == 1]['earn']
earn_female = df[df['male'] == 0]['earn']

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.hist([earn_female, earn_male], bins=20, color=['red', 'blue'], label=['Mujeres', 'Hombres'], alpha=0.7)
plt.title('Distribución de Ingresos por Género')
plt.xlabel('Ingresos (Earn)')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
from scipy import stats

# Filtrar los datos por género
earn_male = df[df['male'] == 1]['earn']
earn_female = df[df['male'] == 0]['earn']

# Realizar una prueba de hipótesis (prueba t de muestras independientes)
t_stat, p_value = stats.ttest_ind(earn_male, earn_female, equal_var=False)

# Nivel de significancia (alfa)
alpha = 0.05

# Comprobar si el valor p es menor que alfa para rechazar la hipótesis nula
if p_value < alpha:
    print("Resultado: Se rechaza la hipótesis nula")
else:
    print("Resultado: No se puede rechazar la hipótesis nula")

# Imprimir el valor t y el valor p
print(f"Valor t: {t_stat}")
print(f"Valor p: {p_value}")
```
