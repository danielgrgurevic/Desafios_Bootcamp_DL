```python
def filtrar_atipicos_iqr(df, columnas):
    """
    Filtra los valores atípicos de las columnas especificadas en un DataFrame utilizando el rango intercuartílico (IQR).

    Parameters:
    df (pd.DataFrame): El DataFrame de entrada.
    columnas (list): Una lista de nombres de columnas para filtrar valores atípicos.

    Returns:
    pd.DataFrame: El DataFrame filtrado sin valores atípicos en las columnas especificadas.
    """
    df_filtrado = df.copy()
    for col in columnas:
        Q1 = df_filtrado[col].quantile(0.25)
        Q3 = df_filtrado[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_filtrado = df_filtrado[(df_filtrado[col] >= lower_bound) & (df_filtrado[col] <= upper_bound)]
    return df_filtrado

columnas_a_filtrar = ['Birth Rate', 'Fertility Rate', 'Life expectancy', 'Physicians per thousand']
df_filtrado = filtrar_atipicos_iqr(df, columnas_a_filtrar)
```

```python
df_filtrado[columnas_a_filtrar].describe()
```
