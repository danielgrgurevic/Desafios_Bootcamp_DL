# 3. Elimina los datos atípicos del dataset. Para las siguientes preguntas, considera el dataset "limpio"

```python
# Seleccionar las columnas numéricas
numeric_columns = df.select_dtypes(include=[np.number]).columns
numeric_columns
```

```python
df_clean = df.copy()

# elimnar outliers con iqr
for column in numeric_columns:
    Q1 = df_clean[column].quantile(0.25)
    Q3 = df_clean[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Eliminar los outliers
    df_clean = df_clean[(df_clean[column] >= lower_bound) & (df_clean[column] <= upper_bound)]

# Verificar el número de filas antes y después de eliminar los outliers
print(f'Número de filas antes de eliminar outliers: {df.shape[0]}')
print(f'Número de filas después de eliminar outliers: {df_clean.shape[0]}')
```

# 6. ¿Es razonable afirmar que, para una persona cualquiera (no necesariamente dentro del dataset), su ritmo cardiaco promedio es 75? Explica y justifica.


Hipótesis nula (H₀): El ritmo cardiaco promedio es 75.

Hipótesis alternativa (H₁): El ritmo cardiaco promedio no es 75.

Dado que el p-value es 0.0507, que es ligeramente mayor que 0.05, no rechazamos la hipótesis nula. Esto significa que no tenemos suficiente evidencia estadística para afirmar que el ritmo cardiaco promedio es diferente de 75. Por lo tanto, es razonable afirmar que el ritmo cardiaco promedio para una persona cualquiera es 75, con base en los datos y el análisis realizado.

```python
import scipy.stats as stats

# Media y desviación estándar del ritmo cardiaco en el dataset
mean_heart_rate = df_clean['heartRate'].mean()
std_heart_rate = df_clean['heartRate'].std()

# Número de muestras
n = df_clean['heartRate'].shape[0]

# Valor hipotético (ritmo cardiaco promedio supuestamente igual a 75)
hypothetical_mean = 75

# Realizar la prueba t para una muestra
t_statistic, p_value = stats.ttest_1samp(df_clean['heartRate'], hypothetical_mean)

print(f"Media del ritmo cardiaco en el dataset: {mean_heart_rate:.2f}")
print(f"Desviación estándar del ritmo cardiaco en el dataset: {std_heart_rate:.2f}")
print(f"t-statistic: {t_statistic:.2f}")
print(f"p-value: {p_value:.4f}")

# Nivel de significancia
alpha = 0.05

if p_value < alpha:
    print("Rechazamos la hipótesis nula. No es razonable afirmar que el ritmo cardiaco promedio es 75.")
else:
    print("No rechazamos la hipótesis nula. Es razonable afirmar que el ritmo cardiaco promedio es 75.")
```

# 7. Ser hombre, ¿influye en el promedio de cigarrillos consumidos por día, dentro de los pacientes fumadores? Explica.


Hipótesis nula (H₀): No hay diferencia en el promedio de cigarrillos consumidos por día entre hombres y mujeres fumadores.

Hipótesis alternativa (H₁): Hay una diferencia en el promedio de cigarrillos consumidos por día entre hombres y mujeres fumadores.

Rechazamos la hipótesis nula: Dado el p-valor muy bajo, podemos rechazar la hipótesis nula de que no hay diferencia en el promedio de cigarrillos consumidos por día entre hombres y mujeres fumadores. Esto significa que hay suficiente evidencia estadística para concluir que ser hombre influye en el promedio de cigarrillos consumidos por día entre los pacientes fumadores.

```python
import scipy.stats as stats

# Filtrar solo los fumadores
smokers = df_clean[df_clean['currentSmoker'] == 1]

# Separar los datos por sexo
male_smokers = smokers[smokers['sex'] == 1]['cigsPerDay']
female_smokers = smokers[smokers['sex'] == 0]['cigsPerDay']

# Calcular la media y desviación estándar para cada grupo
mean_male_smokers = male_smokers.mean()
std_male_smokers = male_smokers.std()
mean_female_smokers = female_smokers.mean()
std_female_smokers = female_smokers.std()

# Realizar la prueba t para dos muestras independientes
t_statistic, p_value = stats.ttest_ind(male_smokers, female_smokers, equal_var=False)

print(f"Media de cigarrillos por día para hombres fumadores: {mean_male_smokers:.2f}")
print(f"Desviación estándar de cigarrillos por día para hombres fumadores: {std_male_smokers:.2f}")
print(f"Media de cigarrillos por día para mujeres fumadoras: {mean_female_smokers:.2f}")
print(f"Desviación estándar de cigarrillos por día para mujeres fumadoras: {std_female_smokers:.2f}")
print(f"t-statistic: {t_statistic:.2f}")
print(f"p-value: {p_value:.4f}")

# Nivel de significancia
alpha = 0.05

if p_value < alpha:
    print("Rechazamos la hipótesis nula. Ser hombre influye en el promedio de cigarrillos consumidos por día entre los pacientes fumadores.")
else:
    print("No rechazamos la hipótesis nula. No podemos concluir que ser hombre influye en el promedio de cigarrillos consumidos por día entre los pacientes fumadores.")
```

# 8. Construye un modelo de regresión lineal que permita relacionar 6 variables del dataset con el índice de masa corporal. (Debes incluir al menos dos variables cualitativas). Evalúa tu modelo y explica

Los coeficientes para las variables sex, currentSmoker, totChol, y sysBP son estadísticamente significativos, indicando una relación significativa con el BMI.
Las variables age y glucose son marginalmente significativas.


El R² de 0.137 indica que el modelo explica una pequeña parte de la variabilidad en el BMI, sugiriendo la influencia de otros factores no incluidos en el modelo



```python
import statsmodels.api as sm

# Definir las variables predictoras (X) y la variable de respuesta (y)
X = df_clean[['age', 'sex', 'currentSmoker', 'totChol', 'sysBP', 'glucose']]
y = df_clean['BMI']

# Agregar una constante al predictor (intercepto)
X = sm.add_constant(X)

# Crear el modelo de regresión lineal
model = sm.OLS(y, X).fit()

# Obtener los resultados del modelo
results = model.summary()

# Imprimir los resultados
print(results)
```
